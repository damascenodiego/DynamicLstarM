```python
## Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# set inline print
%matplotlib inline
```

```python
openssl_tab = "./openssl_releasedates.tab"
df_openssl = pd.read_csv(openssl_tab, sep='\t')
df_openssl["date"] = df_openssl["date"].astype('datetime64[ns]') # set date column as type date
```

```python
ltsdiff_tab = "./ltsdiff.tab"
df_ltsdiff = pd.read_csv(ltsdiff_tab, sep='|')
df_openssldiff = df_ltsdiff[df_ltsdiff["ModelRef"].str.contains("^server_")].replace("\.dot\.kiss$","",regex=True)

# calculating distance in years
from dateutil.relativedelta import relativedelta

dict_versions = df_openssl.set_index("version").to_dict()
print(dict_versions["qsize"])

df_openssldiff["Delta Time"] = df_openssldiff.apply(
        lambda x:
                relativedelta(
                        dict_versions["date"][x["ModelRef"]], 
                        dict_versions["date"][x["ModelUpdt"]]
                ).years,axis=1
        )

df_openssldiff["Delta Size"] = df_openssldiff.apply(
        lambda x: dict_versions["qsize"][x["ModelRef"]]-dict_versions["qsize"][x["ModelUpdt"]])
        ,axis=1
    )

df_openssldiff

# df_openssldiff.apply(lambda x: df_openssl[df_openssl.version == x["ModelRef"]].years, axis=1)


# ax = df_openssl.plot(x ='years', y='qsize', kind = 'scatter')
# plt.rcParams["figure.figsize"] = (10,3)
# plt.show()

# df_openssl["years"]
```
